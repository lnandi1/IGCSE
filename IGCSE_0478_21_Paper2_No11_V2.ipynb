```python
# A  Solution for IGCSE_0478_21_Paper2_No11

# Validates the temperature input. Returns the temperature if valid, otherwise returns None.
def validate_temperature(temp):
    try:
        temp = float(temp)
        if -20.0 <= temp <= 50.0:
            return temp
        else:
            return None
    except ValueError:
        return None

# Converts temperature from Celsius to Fahrenheit and returns the result.
def celsius_to_fahrenheit(temp_celsius):
    return round(temp_celsius * 9/5 + 32, 1)

# Inputs temperatures for a given day and returns a list of temperatures.
def input_temperatures_for_day(day):
    temps = []
    for i in range(24):
        while True:
            temp = input("Enter temperature for hour " + str(i+1) + ": ")
            temp = validate_temperature(temp)
            if temp is not None:
                break
            else:
                print("Invalid input. Temperature must be between -20.0°C and 50.0°C.")
        temps.append(temp)
    return temps

# Calculates the average temperature for a given day and returns the result.
def calculate_average_temperature(day_readings):
    return sum(day_readings) / len(day_readings)

# Calculates the overall average temperature for the whole week and returns the result.
def calculate_overall_average_temperature(weekly_readings):
    total_sum = 0
    for day_readings in weekly_readings:
        total_sum += sum(day_readings)

    total_hours = len(weekly_readings) * 24
    overall_average_temp = total_sum / total_hours

    return overall_average_temp

# Converts a list of temperatures from Celsius to Fahrenheit and returns the result.
def convert_to_fahrenheit(temps):
    fahrenheit_temps = []
    for temp in temps:
        fahrenheit_temp = celsius_to_fahrenheit(temp)
        fahrenheit_temps.append(fahrenheit_temp)
    return fahrenheit_temps

# The main function where the program execution starts.
def main():
    # Initialize and populate the array Days[]
 #   Days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    Days = ["Monday","Tuesday"]

    # Initialize arrays for readings and average temperatures
    Readings = []

    # Input temperatures for each day of the week
    for day in Days:
        print("Input temperatures for", day + ":")
        temps = input_temperatures_for_day(day)
        Readings.append(temps)

    # Calculate average temperature for each day
    AverageTemp = []
    for day_readings in Readings:
        average_temp = calculate_average_temperature(day_readings)
        AverageTemp.append(average_temp)

    # Calculate overall average temperature for the whole week
    overall_average_temp = calculate_overall_average_temperature(Readings)

    # Convert average temperatures to Fahrenheit
    AverageTemp_fahrenheit = convert_to_fahrenheit(AverageTemp)
    overall_average_temp_fahrenheit = celsius_to_fahrenheit(overall_average_temp)

    # Output average temperatures for each day in Celsius and Fahrenheit
    print("\nAverage temperatures for each day:")
    for i in range(len(Days)):
        print(Days[i] + ":", round(AverageTemp[i],1), "C,", round(AverageTemp_fahrenheit[i],1), "F")

    # Output overall average temperature for the whole week in Celsius and Fahrenheit
    print("\nOverall average temperature for the week:", round(overall_average_temp,1), "C,", round(overall_average_temp_fahrenheit,1), "F")

if __name__ == "__main__":
    main()




```
