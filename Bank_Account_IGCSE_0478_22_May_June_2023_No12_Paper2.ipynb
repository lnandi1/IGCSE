```python
# Sample data (replace with your actual data)
Account = [
    ["John Doe", "password1"],
    ["Jane Smith", "password2"],
    # Add more accounts as needed
]

AccDetails = [
    [250.00, 100.00, 200.00],  # Account 0 details: [balance, overdraft limit, withdrawal limit]
    [500.00, 200.00, 300.00],  # Account 1 details
    # Add more account details as needed
]

Size = len(Account)

# Function to check if the account ID exists and if the name and password match
def check_account(account_id, name, password):
    if account_id >= 0 and account_id < Size:
        if Account[account_id][0] == name and Account[account_id][1] == password:
            return True
    return False

# Function to display balance
def display_balance(account_id):
    print(f"Balance: {AccDetails[account_id][0]}")

# Function to withdraw money
def withdraw_money(account_id, amount):
    balance = AccDetails[account_id][0]
    overdraft_limit = AccDetails[account_id][1]
    withdrawal_limit = AccDetails[account_id][2]

    if amount > withdrawal_limit:
        print("Withdrawal amount exceeds withdrawal limit.")
    elif balance - amount < -overdraft_limit:
        print("Withdrawal amount exceeds overdraft limit.")
    else:
        AccDetails[account_id][0] -= amount
        print(f"Withdrawal successful. New balance: {AccDetails[account_id][0]}")

# Function to deposit money
def deposit_money(account_id, amount):
    AccDetails[account_id][0] += amount
    print(f"Deposit successful. New balance: {AccDetails[account_id][0]}")

# Function for the main menu
def main_menu(account_id):
    while True:
        print("\nMenu:")
        print("1. Display balance")
        print("2. Withdraw money")
        print("3. Deposit money")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            display_balance(account_id)
        elif choice == '2':
            amount = float(input("Enter amount to withdraw: "))
            withdraw_money(account_id, amount)
        elif choice == '3':
            amount = float(input("Enter amount to deposit: "))
            deposit_money(account_id, amount)
        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please choose again.")

# Main program
def main():
    account_id = int(input("Enter account ID: "))
    name = input("Enter name: ")
    password = input("Enter password: ")

    if check_account(account_id, name, password):
        print("Login successful.")
        main_menu(account_id)
    else:
        print("Invalid account ID, name, or password.")

if __name__ == "__main__":
    main()
```
