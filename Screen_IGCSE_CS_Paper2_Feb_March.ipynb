```python
# Function to input screen time for each student
def input_screen_time(student_name, days):
    screen_time = []
    for day in range(days):
        time = int(input("Enter screen time for " + student_name + " on day " + str(day + 1) + ": "))
        screen_time.append(time)
    return screen_time

# Function to calculate total screen time for each student
def calculate_total_screen_time(screen_time):
    return sum(screen_time)

# Function to count days with more than 300 minutes of screen time for each student
def count_high_screen_days(screen_time):
    count = 0
    for time in screen_time:
        if time > 300:
            count += 1
    return count

# Function to calculate average weekly screen time for each student
def calculate_average_weekly_screen_time(total_time, days):
    return total_time / days

# Function to calculate average weekly screen time for the whole class
def calculate_average_class_screen_time(total_screen_times, class_size, days):
    return sum(total_screen_times) / (class_size * days)

# Function to find student with the lowest weekly screen time
def find_student_with_lowest_screen_time(total_screen_times):
    min_index = total_screen_times.index(min(total_screen_times))
    return min_index

# Main function
def main():
    # Given data
    StudentName = ["John", "Alice", "Bob"]
    ClassSize = len(StudentName)
    days = 7

    # 2D array to store screen time for each student
    ScreenTime = []

    # Input screen time for each student
    for student in StudentName:
        screen_time = input_screen_time(student, days)
        ScreenTime.append(screen_time)

    # Calculate total screen time, high screen days, and average screen time for the class
    total_screen_times = []
    high_screen_days = []
    for i in range(ClassSize):
        total_time = calculate_total_screen_time(ScreenTime[i])
        high_days = count_high_screen_days(ScreenTime[i])
        average_time = calculate_average_weekly_screen_time(total_time, days)
        total_screen_times.append(total_time)
        high_screen_days.append(high_days)
        print(StudentName[i] + " - Total Screen Time: " + str(total_time // 60) + " hours " + str(total_time % 60) + " minutes, High Screen Days: " + str(high_days))

    # Calculate average weekly screen time for the whole class
    average_class_screen_time = calculate_average_class_screen_time(total_screen_times, ClassSize, days)
    print("Average weekly screen time for the class: " + str(average_class_screen_time) + " minutes")

    # Find student with lowest weekly screen time
    min_index = find_student_with_lowest_screen_time(total_screen_times)
    print("Student with the lowest weekly screen time: " + StudentName[min_index])

# Call main function
if __name__ == "__main__":
    main()
```
