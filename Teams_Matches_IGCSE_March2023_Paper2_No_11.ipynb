# A Solution for Computer Science IGCSE March 2023, Paper 2 - No.11 which involves Teams and Matches #Screenshot 2024-04-17 at 18.13.20.png

```python
# A Solution for Computer Science IGCSE March 2023, Paper 2 - No.11
# Sample data (replace with your actual data)
TeamName = ["Team A", "Team B", "Team C"]
TeamPoints = [
              [3, 2, 1, 3],  # Points for Team A in each match
              [1, 1, 2, 2],  # Points for Team B in each match
              [2, 0, 3, 0]  # Points for Team C in each match
            ]
LeagueSize = len(TeamName) # Here LeagueSize will be 3, 3 Teams
MatchNo = len(TeamPoints[0]) # Here MatchNo will be 4

# Function to calculate total points for each team
def calculate_total_points(TeamPoints):
    total_points = []
    for team in TeamPoints:
        points = sum(team)
        total_points.append(points)
    return total_points # At the end, total_points = [9,6,5]

# Function to count match results for each team
def count_match_results(TeamPoints):
    away_wins = []
    home_wins = []
    draws = []
    losses = []

    for k in range(LeagueSize):
        away_wins.append(0) #So away_wins = [0]
        home_wins.append(0) # So home_wins =[0]
        draws.append(0)     # So draws =[0]
        losses.append(0)    # So losses = [0]

# Reading the elements of the 2D array TeamPoints and seeing if away_win, home_win, draw or loss
    for i in range(MatchNo):
        for j in range(LeagueSize):
            if TeamPoints[j][i] == 3:
                away_wins[j] = away_wins[j] + 1
            elif TeamPoints[j][i] == 2:
                home_wins[j] = home_wins[j] + 1
            elif TeamPoints[j][i] == 1:
                draws[j] = draws[j] + 1
            else:
                losses[j] = losses[j] + 1

    return away_wins, home_wins, draws, losses

# Function to find the team with the highest total points
def find_highest_points_team(TeamName, total_points):
    highest_points = total_points[0]
    highest_team = TeamName[0]

    for i in range(1, len(total_points)):
        if total_points[i] > highest_points:
            highest_points = total_points[i]
            highest_team = TeamName[i]

    return highest_team, highest_points

# Function to find the team with the lowest total points
def find_lowest_points_team(TeamName, total_points):
    lowest_points = total_points[0]
    lowest_team = TeamName[0]

    for i in range(1, len(total_points)):
        if total_points[i] < lowest_points:
            lowest_points = total_points[i]
            lowest_team = TeamName[i]

    return lowest_team, lowest_points

# Function to display results for each team
def display_results(TeamName, total_points, away_wins, home_wins, draws, losses):
    for i in range(LeagueSize):
        print("Team: " + TeamName[i])
        print("Total Points: " + str(total_points[i]))
        print("Match Results:")
        print("Away Wins: " + str(away_wins[i]))
        print("Home Wins: " + str(home_wins[i]))
        print("Draws: " + str(draws[i]))
        print("Losses: " + str(losses[i]))
        print()

# Main function
def main():
    # Calculate total points for each team
    total_points = calculate_total_points(TeamPoints)

    # Count match results for each team
    away_wins, home_wins, draws, losses = count_match_results(TeamPoints)

    # Display results for each team
    display_results(TeamName, total_points, away_wins, home_wins, draws, losses)

    # Find and display the team with the highest and lowest total points
    highest_team, highest_points = find_highest_points_team(TeamName, total_points)
    lowest_team, lowest_points = find_lowest_points_team(TeamName, total_points)
    print("Team with the highest total points: " + highest_team + " (" + str(highest_points) + " points)")
    print("Team with the lowest total points: " + lowest_team + " (" + str(lowest_points) + " points)")
    print("Number of teams in the league: " + str(LeagueSize))
    print("Number of matches played: " + str(MatchNo))

# Run the main function
if __name__ == "__main__":
    main()
```
