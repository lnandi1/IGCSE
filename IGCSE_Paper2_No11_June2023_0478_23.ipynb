```python
# Intialising Variables
contacts =[]
current_size = 0

# Adding New Contact Details
def add_contact(name,telephone):
  global current_size
  contacts.append([name,telephone])
  current_size = current_size + 1
  print("Contact added successfully")


# Display all Contact Details
def display_contacts():
  global current_size
  # Check if no contacts in list
  if current_size == 0:
    print("No contacts to display")
    return
  print("Contacts")
  for contact in contacts:
    print("Name ",contact[0]," ","Telephone ",contact[1])


# Delete All Contact Details
def delete_contacts():
  global current_size, contacts
  contacts = []
  current_size  =0
  print("All contacts successfully deleted")


# Menu to Choose
def menu():
  while True:
    print("\nMenu")
    print("1. Enter new contact details")
    print("2. Display all contact details")
    print("3. Delete all contact details")
    print("4. Exit")

    # Prompt user for Choice
    choice = input("Enter your choice")
    if choice  == "1":

      # Check if the list if full before adding a new contact
      if current_size >= 3: # This 3 should actually be 100
        print("Cannot add more contacts, array is full")
        continue
    # Prompt the user for contact details
      name = input("Enter contact name: Lastname, firstname ")
      telephone = input("Enter a telephone number ")

    # Call the add_contact function
      add_contact(name,telephone)
    # Sort the contacts list alphabetically if there are at least 2 contacts
      if current_size >= 2:
        contacts.sort(key=lambda x:x[0])
    elif choice == "2":
      # Call the display_contacts()
      display_contacts()
    elif choice == "3":
      delete_contacts()
    elif choice == "4":
       # Exit the program
       print("exiting program")
       break
    else:
      # Handle invalid input
      print("Invalid choice. Please choose again ")

# Command to call menu() and start program
menu()












```
